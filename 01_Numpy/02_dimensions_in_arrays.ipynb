```python
import numpy as np
```

```python
#1-D array
array=np.array([1,2,3])
print(array)
shape=np.shape(array)
print(shape)
print(array.ndim)
```

```python
#2-D array
array=np.array([[1,2,3]])
print(array)
shape=np.shape(array)
print(shape)
print(array.ndim)
```

```python
#3-D array
array=np.array([[[1,2,3]]])
print(array)
shape=np.shape(array)
print(shape)
print(array.ndim)
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
print(arr1)
print(arr1.ndim)
```

```python
#NumPy requires equal-length rows so it can create a proper table (array) with a fixed shape
arr2=np.array([[1,2,3],[1,2,3,4]])
print(arr2)
```

```python
arr3=np.array([[[1,2,3],[1,2,3],[1,2,3]]])
print(arr3)
print(arr3.ndim)
```

```python
arr4=np.array([1,2,3], ndmin=5)
print(arr4)
print(arr4.ndim)
```
