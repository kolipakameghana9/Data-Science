```python
import numpy as np
```

```python
#zeros
arr_zero=np.zeros(4)
print(arr_zero)
```

```python
arr_0=np.zeros((3,4))
print(arr_0)
```

```python
#ones
arr_one=np.ones(5)
print(arr_one)
```

```python
arr_1=np.ones((4,5))
print(arr_1)
```

```python
#empty array
arr_em=np.empty(4)
print(arr_em)
```

```python
#range array
arr_rn=np.arange(4)
print(arr_rn)
```

```python
arr_rn=np.arange(1,5)
print(arr_rn)
```

```python
#diagonal
arr_dia=np.eye(3)
print(arr_dia)
```

```python
arr_dia=np.eye(3)
print(arr_dia)
```

```python
#linspace function
arr_lin=np.linspace(1,10,num=5)
print(arr_lin)
```
