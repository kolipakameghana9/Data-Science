```python
import numpy as np
```

insert

```python
var=np.array([1,2,34,5,6,7])
print(var)
x=np.insert(var,2,100)#(variable,index,value)
print(x)
```

```python
var1=var=np.array([[1,2,3,4,6],[21,43,7,89,1],[7,8,9,10,11]])
x=x=np.insert(var1,(1,2),100)#(variable,index,value)
print(x)
```

```python
var1=var=np.array([[1,2,3,4,6],[21,43,7,89,1],[7,8,9,10,11]])
print(var1)
print()
x=x=np.insert(var1,(1,2),100,axis=0)#(variable,index,value,axis)
print(x)
```

delete

```python
var=np.array([1,2,34,5,6,7])
print(var)
x=np.delete(var,2)#(variable,index)
print(x)
```
