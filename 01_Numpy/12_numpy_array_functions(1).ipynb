```python
import numpy as np
```

search array:search an array for a certain value,and return the indexes that get a match

```python
var=np.array([1,2,4,6,8,3,2,6,17,18,16,2])
var1=np.where(var==2)
print(var1)
```

```python
var=np.array([1,2,4,6,8,3,2,6,17,18,16,2])
var1=np.where((var%2)==0)
print(var1)
```

Search sorted array

```python
var=np.array([1,2,3,4,6,7,8,9])
var1=np.searchsorted(var,5)
print(var1)
```

```python
var=np.array([1,2,3,4,6,7,8,9,10,11])
var1=np.searchsorted(var,12,side="right")
print(var1)
```

```python
var=np.array([1,2,3,4,6,7,8,9,10,11])
var1=np.searchsorted(var,[1,3,4])
print(var1)
```

sort array

```python
var=np.array([1,2,3,4,6,21,43,7,89,100,7,8,9,10,11])
var1=np.sort(var)
print(var1)
```

```python
var=np.array(["a","c","a","z","p"])
var1=np.sort(var)
print(var1)
```

```python
var=np.array([[1,2,3,4,6],[21,43,7,89,100],[7,8,9,10,11]])
var1=np.sort(var)
print(var1)
```
