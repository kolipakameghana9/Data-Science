Shape

```python
import numpy as np
arr=np.array([[1,2,3,],[3,4,5]])
print(arr)
print(arr.shape)
```

```python
arr_1=np.array([[1,2,3,4],[5,6,7,8]],ndmin=5)
print(arr_1)
print(arr_1.ndim)
print(arr_1.shape)
```

```python
arr_2=np.array([1,2,3,4,5,6])
x=arr_2.reshape(3,2)
print(x)
print(arr_2.ndim)
print(x.ndim)
```

```python
arr_3=np.array([1,2,3,4,5,6,1,6,9])
x=arr_3.reshape(3,3)
print(x)
print(arr_3.ndim)
print(x.ndim)
```

```python
arr_3=np.array([1,2,3,4,5,6,1,6,9,20,45,10,2,5,8,15,7,8])
x=arr_3.reshape(3,3,2)
print(x)
print(arr_3.ndim)
print(x.ndim)
```
