```python
import numpy as np
```

Arithmetic functions
np.min(x)
np.max(x)
np.argmin(x)
np.argmax(x)
np.sqrt(x)
np.sin(x)
np.cos(x)
np.cumsum(x)


MIN

```python
var=np.array([1,2,3,4,5,6,7])
print(np.min(var))
```

MAX

```python
var=np.array([1,2,3,4,5,6,7])
print(np.max(var))
```

ARGMAX/ARGMIN

```python
var2=np.array([1,2,3,4,5,1,1])
print(np.argmax(var2))
```

```python
var=np.array([1,2,3,4,5,1,1])
print("min:",np.min(var),np.argmin(var))
print("max:",np.max(var),np.argmax(var))
```

```python
var=np.array([[1,2,3],[4,5,1]])
print(np.min(var,axis=0))       #axis=0 ;column
print(np.max(var,axis=1)) #axis=1 ; row
```

SQRT

```python
var1=np.array([4,121,9,256])
print(np.sqrt(var1))
```

SIN

```python
var2=np.array([1,2,3])
print(np.sin(var2))
```

COS

```python
var3=np.array([1,2,3])
print(np.cos(var3))
```

CUMSUM

```python
var4=np.array([1,2,3])
print(np.cumsum(var4))
```
