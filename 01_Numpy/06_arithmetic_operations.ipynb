```python
import numpy as np
```

Arithmetic operations in numpy arrays
a+b-np.add(a,b)
a-b-np.subtract(a,b)
a*b-np.multiply(a,b)
a/b-np.divide(a,b)
a%b-np.mod(a,b)
a**b-np.power(a,b)
1/a-np.reciprocal(a)


```python
var1=np.array([1,2,3,4])
varadd=var1+3
print(varadd)
varadd=np.add(var1,3)
print(varadd)

```

```python
var1=np.array([4,5,6,7])
var2=np.array([10,40,36,72])
varadd=var1+var2
print(varadd)
varadd=np.add(var1,var2)
print(varadd)
```

```python
var1=np.array([4,6,8,4])
varsub=var1-3
print(varsub)
varsub=np.subtract(var1,3)
print(varsub)
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
varsub=var1-var2 
print(varsub)
```

```python
var1=np.array([4,6,8,4])
varmultiply=var1*3
print(varmultiply)
varmultiply=np.multiply(var1,3)
print(varmultiply)
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
varmultiply=var1*var2 
print(varmultiply)
```

```python
var1=np.array([4,6,8,4])
vardivide=var1/3
print(vardivide)
print(np.round(vardivide,2))
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
vardivide=var1/var2 
print(vardivide)
print(np.round(vardivide,2))
```

```python
var1=np.array([4,6,8,4])
varmod=var1%3
print(varmod)
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
varmod=var1%var2 
print(varmod)
```

```python
var1=np.array([4,6,8,4])
varexp=var1**2
print(varexp)
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
varexp=var1**var2 
print(varexp)
```

```python
var1=np.array([4,6,8,4])
varreci=1/var1
print(varreci)
```

```python
var1=np.array([4,6,8,4])
varreci=np.reciprocal(var1)
print(varreci)
```

```python
var1=np.array([2,7,4,9])
var2=np.array([3,6,4,8])
varexp=var1**var2 
print(varexp)
```
