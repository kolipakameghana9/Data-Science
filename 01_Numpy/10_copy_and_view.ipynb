copy vs view

```python
import numpy as np
```

```python
var=np.array([1,2,3,4])
print(var)
print()
co=var.copy()
print("var:",var)
print("copy:",co)

var=np.array([1,2,3,4])
print(var)
print()

co=var.copy()
var[1]=40
print("var:",var)
print("copy:",co)
```

```python
var=np.array([1,2,3,4])
print(var)
print()
view=var.view()
print("var:",var)
print("view:",view)
```

```python
var=np.array([1,2,3,4])
print(var)
print()
view=var.view()
var[1]=20
print("var:",var)
print("view:",view)
```
