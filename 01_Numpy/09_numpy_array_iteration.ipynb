numpy array iteration 

```python
import numpy as np
```

```python
var=np.array([2,5,6,7,8,4,9])
print(var)
print()
for i in var:
     print(var)
```

```python
var=np.array([2,5,6,7,8,4,9])
print(var)
print()
for i in var:
     print(i)
```

```python
var1=np.array([[1,2,3,4],[5,6,7,8]])
print(var1)
print()
for j in var1:
    print(j)
for k in var1:
    for l in k:
        print(l)
```

```python
var1=np.array([[[1,2,3,4],[5,6,7,8]]])
print(var1)
print(var1.ndim)
print()
for i in var1:
    for j in i:
        for k in j:
            print(k)
```

nditer

```python
var1=np.array([[[1,2,3,4],[5,6,7,8]]])
print(var1)
print(var1.ndim)
print()
for i in np.nditer(var1):
    print(i)
```

```python
var2=np.array([[[1,2,3,4],[5,6,7,8]]])
print(var2)
print(var2.ndim)
print()
for i in np.nditer(var2,flags=["buffered"],op_dtypes=["S"]):
    print(i)
```

ndnumerate()

```python
var2=np.array([[[1,2,3,4],[5,6,7,8]]])
print(var2)
print(var2.ndim)
print()
for i,d in np.ndenumerate(var2):
 print(i,d)
```
