stack function

```python
var1=np.array([[1,2,3,4,5],[12,13,14,15,16]])
var2=np.array([[21,22,23,24,25],[32,33,34,35,36]])
print(var1)
print()
print(var2)
var12=np.stack((var1,var2))
print()
print(var12)
```

```python
var1=np.array([[1,2],[12,13]])
var2=np.array([[21,22,],[32,33]])
print()
print(var1)
print()
print(var2)

var12=np.stack((var1,var2),axis=1)
print()
print(var12)
```

```python
var1=np.array([[1,2,3,4,5],[12,13,14,15,16]])
var2=np.array([[21,22,23,24,25],[32,33,34,35,36]])
print(var1)
print()
print(var2)
var12=np.stack((var1,var2))
var13=np.hstack((var1,var2))#row
var14=np.vstack((var1,var2))#column
var15=np.dstack((var1,var2))#height
print()
print("var12:",var12)
print("var13:",var13)
print("var14:",var14)
print(var15)
```

splitting array

```python
var1=np.array([1,2,3,4,5,6])
varsplit=np.array_split(var1,2)
print(varsplit)
print(var1[0])
```

```python
var1=np.array([[1,2,3,4,5],[12,13,14,15,16]])
varsplit=np.array_split(var1,2)
print()
varsplit=np.array_split(var1,2,axis=1)

print(varsplit)
```
