Indexing the numpy array

```python
import numpy as np
```

```python
var=np.array([8,35,7,90,87])
print(var)
print(var[4])
```

```python
var=np.random.randint(3,100,9)
print(var)
print(var[7])
```

```python
var1=np.array([[3,7,10],[8,10,11]])
print(var1)
print(var1[0,1])
```

```python
var1=np.array([[[3,7,10],[8,10,11],[23,32,46]],[[43,81,78],[21,13,11],[3,199,130]]])
print(var1)
print(var1.ndim)
print(var1[1,0,1])
```

Slicing

```python
var=np.array([12,45,3,5,8,2,9])
print(var)
print("1 to 5-",var[1:5])
```

```python
var=np.random.randint(1,100,20)
print(var)
print("5 to end:",var[5:])
print("start to 5:",var[:5])
print("5 to end:",var[5:])
print("start to end:",var[::])
print("2 step:",var[1:18:2])
```

```python
var1=np.array([[1,2,3,36,32],[4,5,6,23,41],[23,1,45,91,12]])
print(var1)
print(var1.ndim)
print(var1[0:1:4])
print(var1[0:1:3])
```
