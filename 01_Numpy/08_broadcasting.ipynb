```python
import numpy as np
```

Broadcasting
conditions to perform operations
-same dimension
(1x3)  (3x1) -->(3x3)

```python
var1=np.array([1,2,3])
var2=np.array([1,2,3,4])
add=np.add(var1,var2)
print(add)
```

```python
var1=np.array([1,2,3])
print(var1)
var2=np.array([[1],[2],[3]])
print(var2)
add=np.add(var1,var2)
print(add)
print(var1.shape)
print(var2.shape)
print(add.shape)
```

```python
a=np.array([[2],[3]])
print(a.shape)
b=np.array([[1,2,3],[1,2,3]])
print(b.shape)
add=np.add(a,b)
print(add)
print(add.shape)
```
