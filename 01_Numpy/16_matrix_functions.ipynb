```python
import numpy as np
```

```python
var=np.matrix([[1,2,3],[4,5,6]])
print(var)
```

Transpose

```python
var=np.matrix([[1,2,3],[4,5,6]])
print(var)
print()
print(np.transpose(var))
```

Swapaxes

```python
var1=np.matrix([[1,2,3],[4,5,6]])
print(var1)
print()
print(np.swapaxes(var1,0,1))
```

```python
var2=np.matrix([[4,5],[7,8]])
print(var2)
print()
print(np.swapaxes(var2,0,1,))
```

Inverse

```python
var3=np.matrix([[4,5],[7,8]])
print(var3)
print()
x=print(np.linalg.inv(var3))
```

power #np.linalg.matrix_power(var,n)

```python
var4=np.matrix([[4,5],[7,8]])
print(var4)
print()
print(np.linalg.matrix_power(var4,3))
```

```python
var4=np.matrix([[4,5],[7,8]])
print(var4)
print()
print(np.linalg.matrix_power(var4,0))
```

```python
var4=np.matrix([[4,5],[7,8]])
print(var4)
print()
print(np.linalg.matrix_power(var4,-1))
```

determinate

```python
var4=np.matrix([[4,5],[7,8]])
print(var4)
print()
print(np.linalg.det(var4))
```
