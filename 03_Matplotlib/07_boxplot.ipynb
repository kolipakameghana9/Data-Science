```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x)
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,notch=True)
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,vert=False)
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,widths=.8)
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True,whis=3)
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True,sym="g+")
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show
```

```python
x=[10,20,30,40,50,120]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="m"))
plt.show
```

```python
x=[10,20,30,40,50,120]
x1=[10,20,40,60,90,120]

plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True,sym="g+")
plt.boxplot(x1,labels=["python"],patch_artist=True,showmeans=True,sym="g+")
plt.show
```

```python
x=[10,20,30,40,50,120]
x1=[10,20,40,60,90,120]
y=[x,x1]
plt.boxplot(y,patch_artist=True,showmeans=True,sym="g+")

plt.show
```
