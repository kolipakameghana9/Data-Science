```python
import matplotlib.pyplot as plt
```

```python
month=[1,2,3,4,5,6,7]
projects=[2,3,1,4,5,3,6]
plt.scatter(month,projects,color="y")
plt.title("scatter plot",fontsize=20)
plt.xlabel("month")
plt.ylabel("projects")
plt.show
```

```python
month=[1,2,3,4,5,6,7]
projects=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","m"]
plt.scatter(month,projects,c=color)
plt.title("scatter plot",fontsize=20)
plt.xlabel("month")
plt.ylabel("projects")
plt.show
```

```python
month=[1,2,3,4,5,6,7]
projects=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","m"]
size=[30,20,50,70,40,90,60]
plt.scatter(month,projects,c=color,s=size)
plt.title("scatter plot",fontsize=20)
plt.xlabel("month")
plt.ylabel("projects")
plt.show
```

```python
month=[1,2,3,4,5,6,7]
projects=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","m"]
size=[30,20,50,70,40,90,60]
plt.scatter(month,projects,c=color,s=size,alpha=.5,marker="*",edgecolor="g",linewidth=2)
plt.title("scatter plot",fontsize=20)
plt.xlabel("month")
plt.ylabel("projects")
plt.show
```

```python
month=[1,2,3,4,5,6,7]
projects=[2,3,1,4,5,3,6]
assignments=[1,5,4,6,7,3,8]
color=[10,40,49,56,23,67,90]
size=[30,20,50,70,40,90,60]

plt.scatter(month,projects,c=color,cmap="BrBG",s=size)
plt.scatter(month,assignments,c=color,cmap="BrBG",s=size)
plt.colorbar()
plt.title("scatter plot",fontsize=20)
plt.xlabel("month")
plt.ylabel("projects")
plt.show
```
