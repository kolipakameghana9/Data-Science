```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
print(x)
y=np.sin(x)
z=np.cos(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data #sin wave
plt.plot(x,y)
plt.show
```

```python
#plotting the data #sin wave
plt.plot(x,z)
plt.show
```

```python
#adding titles
plt.plot(x,y)
plt.xlabel("angle")
plt.xlabel("sin value")
plt.title("sin wave")
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show
```

Bar plot

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=["french","englsih","hindi","arabic","german"]
people=[100,200,250,10,140]
plt.bar(languages,people)
plt.xlabel("languages")
plt.ylabel("number of people")
plt.title("languages vs people")

```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=["french","englsih","hindi","arabic","german"]
people=[100,200,250,10,140]
ax.pie(people,labels=languages,autopct="%1.1f%%")
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y)
ax.scatter(x,z)
plt.show
```
