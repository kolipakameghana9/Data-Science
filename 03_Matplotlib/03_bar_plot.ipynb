```python
x=["python","c","java","cern"]
y=[100,60,80,70]
z=[50,70,80,90]
plt.xlabel("languges")
plt.ylabel("popularity")
plt.title("programmimg languages",fontsize=15)
c=["y","b","g","r"]
plt.bar(x,y,width=.5,color=c,edgecolor="r",label="Bar Plot")
plt.bar(x,z,width=.5,color="m",edgecolor="r",label="Bar Plot")
plt.legend()
plt.show
```

```python
import numpy as np
x=["python","c","java","cern"]
y=[85,70,60,82]
z=[20,30,40,50] 

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languges")
plt.ylabel("popularity")
plt.title("programmimg languages",fontsize=15)


plt.bar(p,y,width,color="r",edgecolor="r",label="Bar Plot")
plt.bar(p1,z,width,color="m",edgecolor="r",label="Bar Plot")
plt.xticks(p+width,x)
plt.legend()
plt.show
```

```python
import numpy as np
x=["python","c","java","cern"]
y=[85,70,60,82]
z=[20,30,40,50] 

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languges")
plt.ylabel("popularity")
plt.title("programmimg languages",fontsize=15)


plt.bar(p,y,width,color="r",edgecolor="r",label="popularity")
plt.bar(p1,z,width,color="m",edgecolor="r",label="popularity1")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show
```

```python
import numpy as np
x=["python","c","java","cern"]
y=[85,70,60,82]
z=[20,30,40,50] 

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languges")
plt.ylabel("popularity")
plt.title("programmimg languages",fontsize=15)


plt.barh(x,y,width,color="r",edgecolor="r",label="popularity")
plt.barh(x,z,width,color="m",edgecolor="r",label="popularity1")

plt.legend()
plt.show
```
