```python
Count plot
A count plot shows how many times each category appears in a dataset.
we use Whenever we have categorical data.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("penguins")
sns.countplot(data=x, x="island")
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.countplot(data=x, x="island",color="r")
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.countplot(data=x, x="island",palette="viridis")
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.countplot(data=x, y="island",hue="sex",palette="viridis")
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.countplot(data=x, x="island",hue="sex",palette="viridis")
plt.figure(figsize=(20,10))
plt.show()
```

A count plot is a Seaborn visualization used to display the frequency (count) of each category in a categorical variable. It helps understand the distribution of categories and compare their counts, making it useful during Exploratory Data Analysis (EDA).
