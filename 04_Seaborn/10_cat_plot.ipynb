```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("penguins")
x
```

```python
sns.catplot(data=x,kind='swarm')
plt.show()
```

```python
sns.catplot(data=x,kind='strip')
plt.show()
```

```python
sns.boxplot(data=x,x="island",y="bill_depth_mm",hue="sex")
plt.show()
```
