```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=sns.load_dataset("penguins")
x
```

```python
corr = x.corr(numeric_only=True)
sns.heatmap(corr,
            annot=True,linewidths=2)
```

```python
corr = x.corr(numeric_only=True)
sns.heatmap(corr,
            annot=True,linewidths=2,linecolor="black",cmap="coolwarm",
    fmt=".2f")
plt.show()
```

```python
corr = x.corr(numeric_only=True)
sns.heatmap(corr,
            annot=True,linewidths=2,linecolor="black",cbar=False,cmap="crest")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

sns.heatmap(data,annot=True,cmap="viridis",linewidths=3)
plt.show()
```

A heatmap is used during Exploratory Data Analysis (EDA) to visualize relationships between features. It helps identify strong positive or negative correlations, detect redundant features, and better understand the dataset before training a machine learning model.
