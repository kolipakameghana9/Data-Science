box plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("penguins")
x
```

```python
sns.boxplot(data=x,x="island",y="bill_length_mm")
plt.show()
```

```python
sns.boxplot(data=x,x="island",y="bill_length_mm",hue="sex")
plt.show()
```

Box: Middle 50% of the data (Q1 to Q3)
Line inside box: Median
Whiskers: Typical range of the data
Dots beyond whiskers: Outliers
