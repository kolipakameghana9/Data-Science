Pair Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("penguins")
x
```

```python
x=sns.load_dataset("penguins")
sns.pairplot(data=x)
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.pairplot(data=x,hue="sex")
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.pairplot(data=x,hue="sex",vars=["body_mass_g","bill_length_mm","bill_depth_mm"])
plt.show()
```

```python
x=sns.load_dataset("penguins")
sns.pairplot(data=x,hue="sex",corner=True)
plt.show()
```

A Pair Plot is a Seaborn visualization that automatically creates scatter plots for every pair of numerical features and displays the distribution of each feature on the diagonal. It is mainly used during Exploratory Data Analysis (EDA) to identify relationships, patterns, clusters, and potential correlations between variables before building a machine learning model.
