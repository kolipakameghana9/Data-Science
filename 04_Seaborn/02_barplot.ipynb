```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x=sns.load_dataset("penguins")
print(x)
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex")
plt.show
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",order=order,hue_order=["Female","Male"],ci=100)
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,hue="sex",order=order,hue_order=["Female","Male"],n_boot=2)
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="bill_length_mm",y="bill_depth_mm",data=x,orient="h")
plt.show()
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,color="g")
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,palette="Accent",hue="sex")
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,palette="pink",hue="sex",saturation=100,err_kws={'color': 'g','linewidth': 10}) #can also use capsize
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,palette="pink",hue="sex",saturation=100,err_kws={'color': 'g','linewidth': 10},dodge=False) #can also use capsize
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=x,palette="pink",hue="sex",saturation=100,err_kws={'color': 'g','linewidth': 10},alpha=.5) #can also use capsize
plt.show
order=["Dream","Torgersen","Biscoe"]
```

```python
sns.set(style="dark")#also use darkgrid
sns.barplot(x="island",y="bill_depth_mm",data=x,palette="pink",hue="sex",saturation=100,err_kws={'color': 'g','linewidth': 10}) 
plt.show
order=["Dream","Torgersen","Biscoe"]
```
