```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#matplotlib
x=[1,2,3,4,5,6,7]
y=[2,5,6,8,9,3,5]
plt.plot(x,y)
plt.show
```

```python
#seaborn line plot
import pandas as pd
x=[1,2,3,4,5,6,7]
y=[2,5,6,8,9,3,5]
z=pd.DataFrame({"x":x,"y":y})
sns.lineplot(x="x",y="y",data=z)
plt.show()

```

```python
x=sns.load_dataset("penguins")
print(x)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex")
plt.show()
```

```python
sns.lineplot(x="body_mass_g",y="sex",data=x)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex",size=10)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex",style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=x,hue="sex",style="sex",palette="binary",markers=["o",">"])
plt.show()
```

```python
y=sns.load_dataset("penguins").head(20)
print(y)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex",style="sex",palette="binary",markers=["o",">"])
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex",style="sex",palette="binary",markers=["o",">"],dashes=False)
plt.grid()
plt.title("line plot")
plt.show()
```
